```python
#Create a dictionary
sdata = {'Ohio': 35000, 'Texas': 54000, 'Oregan': 23780, 'Utah': 34568}

# Move it into a Pandas Series
obj3 = pd.Series(sdata)

obj3
```

```python
# Send back to a Dictionary
obj3.to_dict()
```

```python
states = {'California', 'Ohio','Texas','Oregan'}

#This will override the order of the dictionary keys
obj4 = pd.Series(sdata,index=states)
```

```python
# Note Californa is NaN since there is no sdata for it. Utah is not included since there is no index for it.
print(obj4)
```

```python
pd.isna(obj4)
```

```python
pd.notna(obj4)
```

```python
obj4.isna()
```

```python
obj4.notna()
```

```python
#Returns NaN where labels don't match (missing in either Series)
obj3 + obj4
```

```python
# California: 0 + NaN = NaN (still NaN)
# Ohio: 35000 + 35000 = 70000
# Utah: 34568 + 0 = 34568
obj3.add(obj4, fill_value=0)
```

```python
# Align obj3 to obj4's index, filling missing with 0
obj3_aligned = obj3.reindex(obj4.index, fill_value=0)
obj3_aligned + obj4
```
